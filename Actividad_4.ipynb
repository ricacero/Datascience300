```python
so_excel=so_excel[['Year', 'City', 'Sport', 'Discipline', 'Event', 'Athlete', 'Gender',
       'Country_Code', 'Country', 'Event_gender', 'Medal']]
so_excel
```

```python
indice=list(so_excel.Year)
columnas=so_excel.columns[1:]
valores=so_excel.values[:,1:]
so=pd.DataFrame(valores,index=indice,columns=columnas)
```

```python
so
```

```python
Athlete = so["Athlete"].str.split(',', 1, expand=True)
Athlete.columns = ['last_name', 'first_name']
so = pd.concat([so, Athlete], axis=1)
so
```

```python
so.describe(include='all')
```

```python
so["Country"].value_counts()
```
