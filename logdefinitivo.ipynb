```python
file='20220905-10.log' #Abro el archivo
fp=open(file)
```

```python
contenidoL=fp.readlines() #convierto el archivo en una lista para pder iterar sobre ella
```

```python
#vectores donde alamacenar cada campo
fechae=[]
event=[]
fechac=[]
CallerIDNum=[]
CallerIDName=[]
fechaext=[]
Exten=[]
fechades=[]
DestCallerIDNum=[]
fechaco=[]
ConnectedLineNum=[]
ConnectedLineName=[]
```

```python
#funcion para buscar el campo que queremos
def campo_b (lista,campo,titulo,campo1):
  for j in range (len(lista)): #Se recorre todo el archivo
     if lista[j].startswith(titulo):#si la linea empieza por el campo deseado lo guardo, sino continua
       campo.append(lista[j])
       for x in range(50):#itero hacia atras para detectar la fecha 
         if lista[j-x].startswith('2022'):#si la linea empieza por 2022 osea la fecha lo guardo y freno
          campo1.append(lista[j-x])
          break
         else:
          continue
     else:
      continue
```

```python
#ejecuto la funcion para el campo event
titulo='Event'
campo_b(contenidoL,event,titulo,fechae)
```
